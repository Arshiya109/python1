*   QUESTION NO 1]
*  Define a function, which takes in two strings A and B. Reverses the first string A, adds it to B, and returns the final string.

```python
def reverse_and_concatenate(A, B):
    reversed_A = A[::-1]
    result = reversed_A + B
    return result

string_A = "Hello"
string_B = " World"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```

*   Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
Answer : the female worker that worked the most hours is Penny and she should be paid 18400



```python
def find_female_worker_with_most_hours(work_hours):
    max_hours = 0
    female_worker = None
    for name, hours, gender in work_hours:
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            female_worker = name
    if female_worker:
        pay = max_hours * 20
        return f"The female worker that worked the most hours is {female_worker} and she should be paid ${pay}"
    else:
        return "No female workers found in the list."

work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'), ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]
result = find_female_worker_with_most_hours(work_hours)
print(result)
```

```python
import numpy as np

matrix1 = np.array([[6, 7], [7, 9]])
matrix2 = np.array([[3, 6], [2, 1]])

if matrix1.shape[1] == matrix2.shape[0]:
    dot_product = np.dot(matrix1, matrix2)
    print("Dot Product:")
    print(dot_product)
else:
    print("Matrix dimensions are not compatible for dot product.")
```
